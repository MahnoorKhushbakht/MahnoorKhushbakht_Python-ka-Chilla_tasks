```python
import pandas as pd
bgh=pd.read_csv("Bangladesh.csv")
afg=pd.read_csv('Afghanistan.csv')
```

```python
bgh.drop(['Note'],axis=1,inplace=True)
```

```python
afg.drop(['Note'],axis=1,inplace=True)
```

```python
dat=pd.concat([bgh,afg])
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="ticks", color_codes=True)
p= sns.scatterplot(x="Year",y="Value",hue="Area",data=dat)
plt.show()
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_theme(style="white")
fig = px.pie(dat, values='Value', names='Area')
fig.show()
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd
import numpy as np
sns.set_theme(style="white")
fig = px.scatter(dat, x="Year", y="Value", color="Area",
                 size='Value', hover_data=['Element'])
fig.show()
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd
import numpy as np
fig = px.bar(dat, x='Year', y='Value')
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x='Area',y='Value',data=dat)
plt.title('Population')
plt.show()
```
